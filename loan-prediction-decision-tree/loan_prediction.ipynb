```python
data.drop('LoanID',axis=1,inplace=True,errors='ignore')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat
```

```python
le = LabelEncoder()
```

```python
for i in cat:
    data[i] = le.fit_transform(data[i])
```

```python
data
```

```python
data.isna().sum()
```

```python
X_c = data.drop('Default',axis=1)
y_c = data[['Default']]
```

```python
y_c
```

```python
X_r = data.drop('LoanAmount',axis=1)
y_r = data[['LoanAmount']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
c_X_train,c_X_test,c_y_train,c_y_test = train_test_split(X_c,y_c,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(c_X_train,c_y_train)
```

```python
c_y_pred = DTC.predict(c_X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(c_y_test,c_y_pred)
```

```python
r_X_train,r_X_test,r_y_train,r_y_test = train_test_split(X_r,y_r,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DTR = DecisionTreeRegressor()
```

```python
DTR.fit(r_X_train,r_y_train)
```

```python
r_y_pred = DTR.predict(r_X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(r_y_test,r_y_pred)
```
